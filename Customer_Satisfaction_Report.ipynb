## Checking for nulls and duplicates and handling them.
The data had several nulls but luckily no duplicates.
The nulls were tackled by filling the values in numerical column by mean and mode, and for non- numericals like in Comment we replaced nulls by placing placeholders like 'No Comment'


```python
#checking for nulls!
df.isnull().sum()
```

```python
#Handling nulls!
df = df.fillna(value = {'Age': df['Age'].mean()})
df = df.fillna(value = {'Gender': df['Gender'].mode()[0]})
df = df.fillna(value = {'Location': df['Location'].mode()[0]})
df = df.fillna({'Comment': 'No Comment'})
```

# Performing Sentiment Analysis.
The sentiment analysis was done by using VADER SentimentIntensityAnalyzer which categorized feedback as positive, neutral, or negative.

```python
# Removing Stopwords!
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Example text
text = "This is an example sentence demonstrating stopword removal."

# Tokenize text
words = word_tokenize(text)

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
# Importing Sentiment Analyzer!
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
#Running the polarity score on data!
from tqdm.notebook import tqdm
res = {}
for i, row in tqdm(df.iterrows()):
  text = row['Comment']
  myid = row['CustomerID']
  text = str(row['Comment'])
  res[myid] = sia.polarity_scores(text)
```

```python
# Result of running Sentiment Analyzer
# 0 = neutral, closer to 1 = positve, closer to -1 = negative, compund = avg
res
```

```python
pd.DataFrame(res).T
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns = {'index': 'CustomerID'})
vaders = vaders.merge(df, how = 'left')
```

```python
vaders.head()
```

```python
# Comparing Compound Score by rating to check whether it matches or not!
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.barplot(data = vaders, x = 'Rating', y = 'compound')
ax.set_title('Compound Score by Customer Rating')
plt.show()
```

It verifies that higher the rating, the higher the compund score in positive. It means that Sentiment Analyzer has sentimented correctly.
There is some irregularity because of data as you can see for the 1 value.

```python
category_sentiment = vaders.groupby(['Category', 'compound']).size().unstack().fillna(0)

# Plotting the stacked bar chart
category_sentiment.plot(kind='bar', stacked=True, colormap='Set3', figsize=(10, 6))
plt.title('Sentiment Analysis by Category')
plt.ylabel('Number of Feedback Entries')
plt.tight_layout()
plt.show()
```

The sentiment analysis reveals that:

- **Support** has the highest concentration of positive feedback, with a significant portion of entries falling in the higher sentiment range (0.714 to 0.8685). This indicates a strong positive sentiment in the Support category.
-  **Service, Product, and Delivery** show more mixed sentiments. While there are positive entries, these categories also have a more even distribution across neutral (0.0) and slightly positive sentiment scores (around 0.0946 to 0.5244), indicating varied customer experiences.
- **Negative sentiments** (e.g., -0.3252, -0.3089) are minimal across all categories but are more present in Delivery and Product compared to Support.

Overall, Support stands out as the category with the most positive sentiment, while the other categories have a more balanced mix of sentiments.

```python
# Comparing different ratings!
fig,axes_style = plt.subplots(1,3, figsize = (12,4))
sns.barplot(data = vaders, x = 'Rating', y = 'pos', ax = axes_style[0])
sns.barplot(data = vaders, x = 'Rating', y = 'neu', ax = axes_style[1])
sns.barplot(data = vaders, x = 'Rating', y = 'neg', ax = axes_style[2])
axes_style[0].set_title('Positive')
axes_style[1].set_title('Neutral')
axes_style[2].set_title('Negative')
plt.tight_layout()
plt.show()
```

The visualizations highlight how sentiment correlates with customer ratings:
- Positive Sentiment: Increases with higher ratings, peaking at rating 5.
- Neutral Sentiment: More prevalent at ratings 2, 3, and 4, indicating mixed feelings.
- Negative Sentiment: Highest at rating 3, suggesting dissatisfaction peaks at this mid-level rating.
